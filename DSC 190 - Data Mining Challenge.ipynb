```python
def bool_converter(x):
    if pd.isnull(x):
        return 0
    else:
        return 1
    
train['transit'] = train['transit'].apply(bool_converter)
train['interaction']= train['interaction'].apply(bool_converter)

test['transit'] = test['transit'].apply(bool_converter)
test['interaction']= test['interaction'].apply(bool_converter)


def nan_fill(x):
    if pd.isnull(x):
        return 'none'
    else:
        return x
    
def splitter(x):
    if x=='none':
        return 0
    return len(x.split(','))
```
